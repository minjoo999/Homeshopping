# 문자열 가공

```python
hst3 = hst2.split('\\n')
hst3
```

```python
hst3
```

```python
hst_final = []
# vendor = []
for line in hst3:
    hs = line.split()
    print(hs)
    for i in hs:
        # 짝수번째 요소는 버리고, 홀수번째 요소만 남기기
        if hs.index(i) % 2 == 1:
            hst_final.append(i)

# print(hst_final)
```

```python
hst_final[0].split('(')
```

```python
hst_final[0].replace('(', ' ').replace(')', '')
# hst_final[0].replace(')', '')
```

```python
hh = hst_final[0].replace('(', ' ').replace(')', '')
hh
```

```python
hst_final
```

```python
vendor = []
medicine = []

for hs in hst_final:
    a = hs.replace('(', ' ').replace(')', '')
    hs1 = a.split(' ')
    for i in hs1:
        if hs1.index(i) == 0:
            medicine.append(i)
        else:
            vendor.append(i)

```
